5-. What are iterators in Python and how do they differ from iterables?

-Iterator

An iterator is an object that produces elements of an iterable one at a time.

It remembers its position during iteration.

It has two key methods:

__iter__() → returns the iterator object itself

__next__() → returns the next value, or raises StopIteration when finished.

-Iterable

An iterable is any Python object that can be looped over (like in a for loop).

Examples: list, tuple, string, set, dict, etc.

An iterable has a special method: __iter__() which returns an iterator.

6-Explain the concept of generators in Python and how they are defined.

-A generator is a special type of iterator in Python.

Instead of storing all values in memory at once (like a list), a generator produces values on the fly when you need them.

This makes them memory efficient and useful for handling large data.

-there are two ways to defined gnerators

 (a)Generator Function

Defined like a normal function, but uses the yield keyword instead of return.

Every time yield is called, the function "pauses" and saves its state.

When called again, it resumes from where it left off.


(b)Generator Expression

Similar to list comprehension, but uses parentheses () instead of square brackets [].

7-What are the advantages of using generators over regular functions?

-Advantages of Generators over Regular Functions
1. Memory Efficiency

A regular function that returns a list stores all elements in memory at once.

A generator produces one value at a time using yield, so it doesn’t need to store the whole sequence.

2. Lazy Evaluation

Generators compute values only when needed.

Useful for large datasets or infinite sequences.

3. Improved Performance (Faster for Large Data)

Since values are generated on demand, generators can be faster when dealing with huge data streams.

No time wasted in pre-creating an entire list.

4. Cleaner Code for Iteration

Using yield makes writing iterators simpler compared to creating a full iterator class with __iter__ and __next__.

5. Can Represent Infinite Sequences

Regular functions must stop and return a complete result.

Generators can keep producing values indefinitely (like streaming data).

6. Pipelining (like Unix pipes)

Generators can be chained together for data processing pipelines.

Each stage processes data lazily.
